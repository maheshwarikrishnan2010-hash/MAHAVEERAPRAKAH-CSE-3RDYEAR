# Task 2 & 3: Exploratory Data Analysis & Data Visualization

In this notebook, we'll explore our EV datasets to address **Task 2 (EDA)** and **Task 3 (Data Visualization)**.
- We will inspect data structure, check for anomalies, and validate hypotheses.
- We'll transform this raw data into clear, understandable visual stories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
```

## 1. Data Loading and Initial EDA
Let's load the primary dataset (`ev_dataset.csv`) and the YouTube dataset (`youtube_videos.csv`).

```python
ev_df = pd.read_csv('../data/ev_dataset.csv', low_memory=False)
yt_df = pd.read_csv('../data/youtube_videos.csv')

print("EV Dataset Shape:", ev_df.shape)
print("YouTube Dataset Shape:", yt_df.shape)
```

### Asking Questions: What is the structure of the data? Are there missing values?

```python
print("--- Columns & Data Types ---")
print(ev_df.info())

print("\n--- Missing Values ---")
print(ev_df.isnull().sum()[ev_df.isnull().sum() > 0])
```

## 2. Visualizing Trends and Patterns
Let's see which EV manufacturers dominate the dataset.

```python
make_counts = ev_df['make'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=make_counts.values, y=make_counts.index, palette='viridis')
plt.title('Top 10 EV Makes Registered', fontsize=14)
plt.xlabel('Total Registrations', fontsize=12)
plt.ylabel('Vehicle Make', fontsize=12)
plt.tight_layout()
plt.show()
```

## 3. Testing Hypotheses
**Hypothesis:** Has the average electric range of Battery Electric Vehicles (BEVs) improved significantly over recent model years?

```python
if 'ev_type' in ev_df.columns and 'electric_range' in ev_df.columns:
    # Filter out plug-in hybrids to focus on pure EVs
    bev_df = ev_df[ev_df['ev_type'].str.contains('Battery', na=False, case=False)]
    
    # Exclude ranges of 0 which often signify missing data in this specific dataset
    bev_df = bev_df[bev_df['electric_range'] > 0]
    
    yearly_range = bev_df.groupby('model_year')['electric_range'].mean().reset_index()
    yearly_range = yearly_range[(yearly_range['model_year'] >= 2010) & (yearly_range['model_year'] <= 2024)]
    
    plt.figure(figsize=(10, 5))
    sns.regplot(data=yearly_range, x='model_year', y='electric_range', marker='o', color='teal')
    plt.title('Average Electric Range of BEVs by Model Year (2010-2024)', fontsize=14)
    plt.xlabel('Model Year', fontsize=12)
    plt.ylabel('Avg Range (Miles)', fontsize=12)
    plt.xticks(np.arange(2010, 2025, 2))
    plt.tight_layout()
    plt.show()
```

**Observation:** The regression line demonstrates a positive trend, confirming our hypothesis that battery capacity and efficiency have consistently improved over the years.

## 4. Visualizing Engagement Data (YouTube)
Let's see if there is a strong correlation between how many views an EV video gets and its likes.

```python
if 'views' in yt_df.columns and 'likes' in yt_df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=yt_df, x='views', y='likes', alpha=0.5, color='coral', edgecolor=None)
    plt.title('Correlation between Video Views and Likes', fontsize=14)
    plt.xlabel('Views (Millions)', fontsize=12)
    plt.ylabel('Likes', fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # Print the correlation coefficient
    corr = yt_df['views'].corr(yt_df['likes'])
    print(f"Pearson Correlation Coefficient: {corr:.3f}")
```
