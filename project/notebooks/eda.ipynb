# EV & YouTube Data Exploratory Data Analysis
This notebook visualizes and explores the data quality and patterns in our EV datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid")
data_dir = "../data"
ev_df = pd.read_csv(os.path.join(data_dir, "ev_dataset.csv"))
yt_df = pd.read_csv(os.path.join(data_dir, "youtube_videos.csv"))
```

```python
# --- EV Data Quality ---
print("EV Dataset Shape:", ev_df.shape)
print("Missing values in EV Data:")
print(ev_df.isnull().sum()[ev_df.isnull().sum() > 0])
```

```python
# --- EV Makes Distribution ---
top_makes = ev_df["make"].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_makes.values, y=top_makes.index, palette="viridis")
plt.title("Top 10 EV Makes")
plt.show()
```

```python
# --- YouTube EDA ---
print("YouTube Data Shape:", yt_df.shape)
plt.figure(figsize=(8,5))
sns.scatterplot(data=yt_df, x="views", y="likes", alpha=0.6, color="red")
plt.title("Views vs Likes for EV Videos")
plt.show()
```
